```python
import matplotlib,pylab as plt
from sklearn import linear_model

reg = linear_model.LinearRegression()

X = [[174], [152], [138], [128], [186]]
y = [71, 55, 46, 38, 88]
reg.fit(X, y)

print(reg.predict([[165]]))

plt.scatter(X, y, color = 'black')
plt.scatter(165, reg.predict([[165]]), color = 'red')
y_pred = reg.predict(X)

plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()
```

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
print(iris)
print(iris.DESCR)

print(iris.data)
print(iris.target_names)

df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df['target'] = iris.target

df['target'] = df['target'].map({0 : "setosa", 1 : "versicolor", 2:"virginica"})
print(df)

x_data = df.iloc[:, :-1]
y_data = df.iloc[:, [-1]]

sns.pairplot(df, x_vars = ["sepal length (cm)"], y_vars = ["sepal width (cm)"], hue = "target", height = 5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state = 0)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

plt.scatter(X[:,0], X[:,1], c=y , cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize = 18)
plt.ylabel('Sepal Width', fontsize = 18)

km = KMeans(n_clusters = 3, n_jobs = 4, random_state = 21)
km.fit(X)

centers = km.cluster_centers_
print(centers)

plt.scatter(centers[0, 0], centers[0, 1], color = 'blue')
plt.scatter(centers[1, 0], centers[1, 1], color = 'green')
plt.scatter(centers[2, 0], centers[2, 1], color = 'black')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state = 0)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

classes = {0:'setosa', 1:'versicolor', 2:'virginica'}

x_new = [[3, 4, 5, 2],
         [5, 4, 2, 2]]
y_predict = knn.predict(x_new)

print(classes[y_predict[0]])
print(classes[y_predict[1]])
```
